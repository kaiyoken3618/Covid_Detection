```python
import pandas as pd
import os
import numpy as np 
from google.colab import drive
drive.mount('/content/drive')
```

```python
datapath1='/content/drive/My Drive/Covid_Detect/covid-chestxray-dataset-master/covid-chestxray-dataset-master'
dataset_path='/content/drive/My Drive/Covid_Detect/dataset'

categories=os.listdir(dataset_path)
print(categories)

dataset=pd.read_csv(os.path.join(datapath1,'/content/drive/MyDrive/Covid_Detect/covid-chestxray-dataset-master/covid-chestxray-dataset-master/metadata.csv'))
findings=dataset['finding']
image_names=dataset['filename']
```

```python
positives_index=np.where(findings=='Pneumonia/Viral/COVID-19')[0]
positive_image_names=image_names[positives_index]
```

```python
import cv2

for positive_image_name in positive_image_names:
    image=cv2.imread(os.path.join(datapath1,'/content/drive/MyDrive/Covid_Detect/covid-chestxray-dataset-master/covid-chestxray-dataset-master/images',positive_image_name))
    try:
        cv2.imwrite(os.path.join(dataset_path,categories[1],positive_image_name),image)
    except Exception as e:
        print(e)
```

```python
datapath2='/content/drive/My Drive/Covid_Detect/archive'

dataset=pd.read_csv(os.path.join(datapath2,'/content/drive/MyDrive/Covid_Detect/archive/Chest_xray_Corona_Metadata.csv'))
findings=dataset['Label']
image_names=dataset['X_ray_image_name']
```

```python
negative_index=np.where(findings=='Normal')[0]
```

```python
negative_image_names=image_names[negative_index]
```

```python
for negative_image_name in negative_image_names:
    image=cv2.imread(os.path.join(datapath2,'/content/drive/MyDrive/Covid_Detect/archive/Coronahack-Chest-XRay-Dataset/Coronahack-Chest-XRay-Dataset/train',negative_image_name))
    try:
        cv2.imwrite(os.path.join(dataset_path,categories[0],negative_image_name),image)
    except Exception as e:
        print(e)
```
